# 🎼 VIRASAT AI — GPU Fast Separator (Colab & Kaggle)

This standalone notebook utilizes the free GPUs on Google Colab or Kaggle to instantly download any heritage song from YouTube, isolate the vocals using Demucs v4 (htdemucs_ft), and prepare the clean vocal track for you to download.


```python
print("Installing the latest Demucs AI and YT-Downloader...")
!pip install -qU demucs yt-dlp
import torch
print("\n✅ Installation Complete!")
print(f"GPU Accelerated Processing Active: {torch.cuda.is_available()}")
```

### ⬇️ 1. Enter the YouTube link of the heritage song below and run the cell:

```python
youtube_url = "https://www.youtube.com/watch?v=S216B9Gihg0"  # Example: Ranjish Hi Sahi - Mehdi Hassan

import os
os.makedirs("raw_audio", exist_ok=True)
print("Downloading high-quality audio...")
!yt-dlp -x --audio-format wav --audio-quality 0 -o "raw_audio/song.%(ext)s" {youtube_url}
print("\n✅ Download Complete!")
```

### 🧠 2. Run the AI Separator (Extracts Vocals using GPU)

```python
# We use the fine-tuned model (htdemucs_ft) which is better for old tracks.
import time
start_time = time.time()
print("Running Demucs AI... This will take a moment depending on the GPU.")

!demucs -n htdemucs_ft --two-stems vocals raw_audio/song.wav -o separated_audio

end_time = time.time()
print(f"\n✅ Extraction Complete in {round(end_time - start_time)} seconds!")
```

### 💾 3. Zip and Download the Clean Vocals

```python
import shutil
try:
    from google.colab import files
    # Zip the vocals folder
    shutil.make_archive('virasat_clean_vocals', 'zip', 'separated_audio/htdemucs_ft/song')
    print("Zipping complete! Downloading to your computer now...")
    files.download('virasat_clean_vocals.zip')
except ImportError:
    # If running on Kaggle
    shutil.make_archive('/kaggle/working/virasat_clean_vocals', 'zip', 'separated_audio/htdemucs_ft/song')
    print("Zipping complete! Find 'virasat_clean_vocals.zip' in the Kaggle Output section on the right side of the screen to download.")
```
